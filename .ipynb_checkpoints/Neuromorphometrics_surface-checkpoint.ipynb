## Load labels_Neuromorphometrics, create WM surface, save as obj

```python
import numpy as np
from skimage.measure import marching_cubes
import scipy.ndimage as ndi
import nibabel as nib
from collections import defaultdict

# Function to smooth the mesh using Laplacian smoothing (optimized version)
def laplacian_smooth_optimized(vertices, faces, iterations=10, lambda_=0.5):
    vertex_count = vertices.shape[0]
    
    # Create a dictionary for vertex neighbors
    vertex_neighbors = defaultdict(list)
    
    # Populate the vertex neighbors by iterating through faces
    for face in faces:
        for i in range(3):
            for j in range(i + 1, 3):
                vertex_neighbors[face[i]].append(face[j])
                vertex_neighbors[face[j]].append(face[i])

    # Convert to numpy for faster processing
    vertex_neighbors = {key: np.array(value) for key, value in vertex_neighbors.items()}

    # Smoothing iteration
    smoothed_vertices = np.copy(vertices)
    
    for _ in range(iterations):
        new_vertices = np.copy(smoothed_vertices)
        
        for vertex_idx in range(vertex_count):
            # Get neighbors
            neighbors = vertex_neighbors.get(vertex_idx, [])
            
            # If there are neighbors, average the neighbors' positions
            if len(neighbors) > 0:
                neighbor_positions = smoothed_vertices[neighbors]
                avg_position = np.mean(neighbor_positions, axis=0)
                
                # Perform the Laplacian update
                new_vertices[vertex_idx] = lambda_ * avg_position + (1 - lambda_) * smoothed_vertices[vertex_idx]
        
        smoothed_vertices = new_vertices
    
    return smoothed_vertices

# Step 1: Load the original NIfTI file
nii_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/labels_Neuromorphometrics.nii"
nii_img = nib.load(nii_path)
nii_data = nii_img.get_fdata()

# Step 2: Isolate White Matter (WM) regions (right = 44, left = 45)
wm_right_mask = (nii_data == 44).astype(np.uint8)
wm_left_mask = (nii_data == 45).astype(np.uint8)

# Step 3: Dilate each WM mask by one voxel
dilated_wm_right_mask = ndi.binary_dilation(wm_right_mask, structure=np.ones((3, 3, 3))).astype(np.uint8)
dilated_wm_left_mask = ndi.binary_dilation(wm_left_mask, structure=np.ones((3, 3, 3))).astype(np.uint8)

# Step 4: Extract surfaces using marching cubes (one for each hemisphere)
def extract_surface(mask):
    vertices, faces, _, _ = marching_cubes(mask, level=0)
    return vertices, faces

right_vertices, right_faces = extract_surface(dilated_wm_right_mask)
left_vertices, left_faces = extract_surface(dilated_wm_left_mask)

# Step 5: Smooth the surfaces using the optimized Laplacian smoothing function
right_smoothed_vertices = laplacian_smooth_optimized(right_vertices, right_faces, iterations=10, lambda_=0.5)
left_smoothed_vertices = laplacian_smooth_optimized(left_vertices, left_faces, iterations=10, lambda_=0.5)

# Step 6: Get the affine transformation from the original NIfTI file (to ensure correct coordinates)
affine = nii_img.affine

# Step 7: Convert smoothed vertices (voxel coordinates) back to world coordinates
def convert_to_world_coordinates(vertices, affine):
    # Apply the affine matrix (inverse) to convert from voxel to world coordinates
    vertices_world = np.dot(vertices, affine[:3, :3].T) + affine[:3, 3]
    return vertices_world

right_smoothed_vertices_world = convert_to_world_coordinates(right_smoothed_vertices, affine)
left_smoothed_vertices_world = convert_to_world_coordinates(left_smoothed_vertices, affine)

# Step 8: Save the smoothed surfaces to .obj files
def save_obj(filename, vertices, faces):
    with open(filename, 'w') as f:
        # Write the vertices in world coordinates
        for v in vertices:
            f.write(f"v {v[0]} {v[1]} {v[2]}\n")
        
        # Write the faces (1-based indexing for OBJ format)
        for face in faces:
            f.write(f"f {face[0]+1} {face[1]+1} {face[2]+1}\n")

# Output file paths for the surfaces
output_right_obj_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_right_surface_smoothed.obj"
output_left_obj_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_left_surface_smoothed.obj"

# Save the smoothed surfaces for both hemispheres
save_obj(output_right_obj_path, right_smoothed_vertices_world, right_faces)
save_obj(output_left_obj_path, left_smoothed_vertices_world, left_faces)

print(f"Right WM surface saved as {output_right_obj_path}")
print(f"Left WM surface saved as {output_left_obj_path}")
```

## Load labels_Neuromorphometrics, create WM surface, save as gii

```python
import numpy as np
from skimage.measure import marching_cubes
import scipy.ndimage as ndi
import nibabel as nib
from collections import defaultdict

# Function to smooth the mesh using Laplacian smoothing (optimized version)
def laplacian_smooth_optimized(vertices, faces, iterations=10, lambda_=0.5):
    vertex_count = vertices.shape[0]
    
    # Create a dictionary for vertex neighbors
    vertex_neighbors = defaultdict(list)
    
    # Populate the vertex neighbors by iterating through faces
    for face in faces:
        for i in range(3):
            for j in range(i + 1, 3):
                vertex_neighbors[face[i]].append(face[j])
                vertex_neighbors[face[j]].append(face[i])

    # Convert to numpy for faster processing
    vertex_neighbors = {key: np.array(value) for key, value in vertex_neighbors.items()}

    # Smoothing iteration
    smoothed_vertices = np.copy(vertices)
    
    for _ in range(iterations):
        new_vertices = np.copy(smoothed_vertices)
        
        for vertex_idx in range(vertex_count):
            # Get neighbors
            neighbors = vertex_neighbors.get(vertex_idx, [])
            
            # If there are neighbors, average the neighbors' positions
            if len(neighbors) > 0:
                neighbor_positions = smoothed_vertices[neighbors]
                avg_position = np.mean(neighbor_positions, axis=0)
                
                # Perform the Laplacian update
                new_vertices[vertex_idx] = lambda_ * avg_position + (1 - lambda_) * smoothed_vertices[vertex_idx]
        
        smoothed_vertices = new_vertices
    
    return smoothed_vertices

# Step 1: Load the original NIfTI file
nii_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/labels_Neuromorphometrics.nii"
nii_img = nib.load(nii_path)
nii_data = nii_img.get_fdata()

# Step 2: Isolate White Matter (WM) regions (right = 44, left = 45)
wm_right_mask = (nii_data == 44).astype(np.uint8)
wm_left_mask = (nii_data == 45).astype(np.uint8)

# Step 3: Dilate each WM mask by one voxel
dilated_wm_right_mask = ndi.binary_dilation(wm_right_mask, structure=np.ones((3, 3, 3))).astype(np.uint8)
dilated_wm_left_mask = ndi.binary_dilation(wm_left_mask, structure=np.ones((3, 3, 3))).astype(np.uint8)

# Step 4: Extract surfaces using marching cubes (one for each hemisphere)
def extract_surface(mask):
    vertices, faces, _, _ = marching_cubes(mask, level=0)
    return vertices, faces

right_vertices, right_faces = extract_surface(dilated_wm_right_mask)
left_vertices, left_faces = extract_surface(dilated_wm_left_mask)

# Step 5: Smooth the surfaces using the optimized Laplacian smoothing function
right_smoothed_vertices = laplacian_smooth_optimized(right_vertices, right_faces, iterations=10, lambda_=0.5)
left_smoothed_vertices = laplacian_smooth_optimized(left_vertices, left_faces, iterations=10, lambda_=0.5)

# Step 6: Get the affine transformation from the original NIfTI file (to ensure correct coordinates)
affine = nii_img.affine

# Step 7: Convert smoothed vertices (voxel coordinates) back to world coordinates
def convert_to_world_coordinates(vertices, affine):
    # Apply the affine matrix (inverse) to convert from voxel to world coordinates
    vertices_world = np.dot(vertices, affine[:3, :3].T) + affine[:3, 3]
    return vertices_world

right_smoothed_vertices_world = convert_to_world_coordinates(right_smoothed_vertices, affine)
left_smoothed_vertices_world = convert_to_world_coordinates(left_smoothed_vertices, affine)

from nibabel.gifti import GiftiImage, GiftiDataArray

def save_gifti_surface(filename, vertices, faces):
    # Create GiftiDataArrays
    coord_array = GiftiDataArray(data=vertices.astype(np.float32), intent='NIFTI_INTENT_POINTSET')
    face_array = GiftiDataArray(data=faces.astype(np.int32), intent='NIFTI_INTENT_TRIANGLE')

    # Create the Gifti image
    gii_img = GiftiImage(darrays=[coord_array, face_array])

    # Save the image
    nib.save(gii_img, filename)

def combine_hemispheres_and_save(filename, left_vertices, left_faces, right_vertices, right_faces):
    # Offset the right faces by the number of left vertices
    offset = left_vertices.shape[0]
    right_faces_offset = right_faces + offset

    # Combine vertices and faces
    combined_vertices = np.vstack((left_vertices, right_vertices))
    combined_faces = np.vstack((left_faces, right_faces_offset))

    # Create GIFTI arrays
    coord_array = GiftiDataArray(data=combined_vertices.astype(np.float32), intent='NIFTI_INTENT_POINTSET')
    face_array = GiftiDataArray(data=combined_faces.astype(np.int32), intent='NIFTI_INTENT_TRIANGLE')

    # Save GIFTI
    gii_img = GiftiImage(darrays=[coord_array, face_array])
    nib.save(gii_img, filename)


# Output file paths for the surfaces
output_right_gii_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_right_surface_smoothed.gii"
output_left_gii_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_left_surface_smoothed.gii"

# Save the smoothed surfaces for both hemispheres
save_gifti_surface(output_right_gii_path, right_smoothed_vertices_world, right_faces)
save_gifti_surface(output_left_gii_path, left_smoothed_vertices_world, left_faces)

print(f"Right WM surface saved as {output_right_gii_path}")
print(f"Left WM surface saved as {output_left_gii_path}")

combined_gii_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_combined_surface_smoothed.gii"
combine_hemispheres_and_save(combined_gii_path, left_smoothed_vertices_world, left_faces, right_smoothed_vertices_world, right_faces)

print(f"Combined WM surface saved as {combined_gii_path}")
```

## Load cortical_only, extract label intersection surface

```python
import numpy as np
import nibabel as nib
from collections import Counter
from scipy.spatial.distance import cdist
from scipy.ndimage import distance_transform_edt

# Load the neuromorphometric labels (cortical only mask)
cortical_mask_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/labels_Neuromorphometrics_cortical_only.nii.gz"  # Path to the cortical mask
nii_img_cortical = nib.load(cortical_mask_path)
cortical_data = nii_img_cortical.get_fdata()

# Load an OBJ file (surface) into vertices and faces
def load_obj(obj_path):
    vertices = []
    faces = []
    with open(obj_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if parts[0] == 'v':  # Vertex line
                vertices.append([float(x) for x in parts[1:]])
            elif parts[0] == 'f':  # Face line
                faces.append([int(x.split('/')[0]) - 1 for x in parts[1:]])  # OBJ format is 1-indexed, we need 0-indexed
    return np.array(vertices), np.array(faces)

# Function to get the majority label from the 3x3x3 neighborhood around a voxel
def get_majority_label(voxel, image_data, neighborhood_size=3):
    x, y, z = voxel
    half_size = neighborhood_size // 2

    # Get the coordinates of the neighborhood (in 3x3x3)
    neighborhood = image_data[max(0, x - half_size):min(image_data.shape[0], x + half_size + 1),
                              max(0, y - half_size):min(image_data.shape[1], y + half_size + 1),
                              max(0, z - half_size):min(image_data.shape[2], z + half_size + 1)]

    # Flatten the neighborhood and count the most frequent label (majority vote)
    labels = neighborhood.flatten()
    
    # Count only non-zero labels
    non_zero_labels = [label for label in labels if label > 0]
    
    if len(non_zero_labels) > 0:
        majority_label = Counter(non_zero_labels).most_common(1)[0][0]  # Majority label
        return majority_label
    else:
        return 0  # If there are no non-zero labels, return 0

# Function to sample the labels for all vertices on the surface
def sample_labels_for_surface(vertices, cortical_data, affine, neighborhood_size=3):
    sampled_labels = []
    for vertex in vertices:
        # Convert vertex to voxel coordinates using the affine transformation
        voxel = np.round(np.linalg.inv(affine) @ np.append(vertex, 1))[:3].astype(int)

        # Check if the voxel is within bounds
        if np.all(voxel >= 0) and np.all(voxel < cortical_data.shape):
            # Sample the label using the majority vote from the neighborhood
            label = get_majority_label(voxel, cortical_data, neighborhood_size)
        else:
            label = 0  # If out of bounds, set the label to 0 (or handle this case differently)
        sampled_labels.append(label)
    return np.array(sampled_labels)

# Function to replace zero labels with the nearest non-zero label using distance transform
def replace_zero_labels_with_nearest(labels, vertices, cortical_data, affine, neighborhood_size=3):
    # For each vertex with label 0, check its neighborhood and replace the 0 if necessary
    for idx, label in enumerate(labels):
        if label == 0:
            voxel = np.round(np.linalg.inv(affine) @ np.append(vertices[idx], 1))[:3].astype(int)
            
            # Check if the voxel is within bounds
            if np.all(voxel >= 0) and np.all(voxel < cortical_data.shape):
                # Get the majority label from the 3x3 neighborhood
                labels[idx] = get_majority_label(voxel, cortical_data, neighborhood_size)
    return labels

# Function to save the sampled labels to a text file
def save_labels_to_txt(labels, output_txt_path):
    np.savetxt(output_txt_path, labels, fmt="%d")
    print(f"Labels saved to {output_txt_path}")

# Example usage for left and right hemisphere surfaces
# Load the right and left hemisphere surfaces
right_surface_vertices, right_surface_faces = load_obj("/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_right_surface_smoothed.obj")
left_surface_vertices, left_surface_faces = load_obj("/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_left_surface_smoothed.obj")

# Get the affine transformation from the original cortical NIfTI file
affine = nii_img_cortical.affine

# Sample the labels for both right and left surfaces
right_sampled_labels = sample_labels_for_surface(right_surface_vertices, cortical_data, affine, neighborhood_size=3)
left_sampled_labels = sample_labels_for_surface(left_surface_vertices, cortical_data, affine, neighborhood_size=3)

# Replace the zero labels with the nearest non-zero labels
right_sampled_labels = replace_zero_labels_with_nearest(right_sampled_labels, right_surface_vertices, cortical_data, affine, neighborhood_size=3)
left_sampled_labels = replace_zero_labels_with_nearest(left_sampled_labels, left_surface_vertices, cortical_data, affine, neighborhood_size=3)

# Save the sampled labels to text files
right_labels_txt_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/right_surface_labels.txt"
left_labels_txt_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/left_surface_labels.txt"

save_labels_to_txt(right_sampled_labels, right_labels_txt_path)
save_labels_to_txt(left_sampled_labels, left_labels_txt_path)
```

## Find boundary vertices

```python
import numpy as np
import nibabel as nib
from collections import Counter

# ========== Load Surface ==========
def load_obj(obj_path):
    vertices = []
    faces = []
    with open(obj_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if parts[0] == 'v':
                vertices.append([float(x) for x in parts[1:]])
            elif parts[0] == 'f':
                faces.append([int(x.split('/')[0]) - 1 for x in parts[1:]])
    return np.array(vertices), np.array(faces)

# ========== Neighbor & Boundary Detection ==========
def build_vertex_neighbors(faces, n_vertices):
    neighbors = [[] for _ in range(n_vertices)]
    for tri in faces:
        for i in range(3):
            a, b = tri[i], tri[(i+1)%3]
            if b not in neighbors[a]:
                neighbors[a].append(b)
            if a not in neighbors[b]:
                neighbors[b].append(a)
    return neighbors

def find_label_boundaries(vertices, faces, labels):
    n_vertices = len(vertices)
    neighbors = build_vertex_neighbors(faces, n_vertices)
    boundary_mask = np.zeros(n_vertices, dtype=bool)
    for i in range(n_vertices):
        my_label = labels[i]
        for n in neighbors[i]:
            if labels[n] != my_label:
                boundary_mask[i] = True
                break
    return boundary_mask

# ========== Smoothing ==========
def smooth_labels_at_boundaries(vertices, labels, faces, n_iter=1):
    smoothed_labels = labels.copy()
    n_vertices = len(labels)
    neighbors = build_vertex_neighbors(faces, n_vertices)

    for _ in range(n_iter):
        boundary_mask = find_label_boundaries(vertices, faces, smoothed_labels)
        new_labels = smoothed_labels.copy()
        for idx in np.where(boundary_mask)[0]:
            neighbor_labels = [smoothed_labels[n] for n in neighbors[idx] if smoothed_labels[n] > 0]
            if neighbor_labels:
                most_common = Counter(neighbor_labels).most_common(1)[0][0]
                new_labels[idx] = most_common
        smoothed_labels = new_labels

    return smoothed_labels

# ========== Save ==========
def save_labels_to_txt(labels, path):
    np.savetxt(path, labels, fmt="%d")
    print(f"Saved: {path}")

def save_boundary_mask(mask, path):
    np.savetxt(path, mask.astype(int), fmt="%d")
    print(f"Boundary mask saved: {path}")

# ========== Paths ==========
right_surface_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_right_surface_smoothed.obj"
left_surface_path  = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_left_surface_smoothed.obj"
right_labels_path  = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/right_surface_labels.txt"
left_labels_path   = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/left_surface_labels.txt"

# ========== Load ==========
right_vertices, right_faces = load_obj(right_surface_path)
left_vertices, left_faces = load_obj(left_surface_path)
right_labels = np.loadtxt(right_labels_path, dtype=int)
left_labels = np.loadtxt(left_labels_path, dtype=int)

# ========== Optional: Save boundary mask for visualization ==========
right_boundary = find_label_boundaries(right_vertices, right_faces, right_labels)
left_boundary  = find_label_boundaries(left_vertices, left_faces, left_labels)
save_boundary_mask(right_boundary, "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/right_boundary_mask.txt")
save_boundary_mask(left_boundary,  "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/left_boundary_mask.txt")
```

```python
import numpy as np
import plotly.graph_objects as go

# Function to load vertices from an .obj file
def load_obj_vertices(obj_path):
    vertices = []
    with open(obj_path, 'r') as f:
        for line in f:
            if line.startswith("v "):  # Vertex line
                parts = line.split()
                vertices.append([float(parts[1]), float(parts[2]), float(parts[3])])
    return np.array(vertices)

# Function to create and save 3D scatter plot
def create_3d_plot(vertices, labels, hemisphere, output_path):
    # Create a 3D scatter plot for the vertices with labels
    scatter = go.Scatter3d(
        x=vertices[:, 0],
        y=vertices[:, 1],
        z=vertices[:, 2],
        mode='markers',
        marker=dict(
            size=5.5,  # Increase the size of the dots
            color=labels,  # Color by the labels
            colorscale='Jet',  # Change color map if needed
            opacity=1.0,
            colorbar=dict(title='')
        )
    )

    # Create the layout for the 3D plot
    
    layout = go.Layout(
        scene=dict(
            xaxis=dict(
                title='',
                showgrid=False,
                zeroline=False,
                showticklabels=False,
                backgroundcolor='white',  # Remove grey plane
                showbackground=False
            ),
            yaxis=dict(
                title='',
                showgrid=False,
                zeroline=False,
                showticklabels=False,
                backgroundcolor='white',  # Remove grey plane
                showbackground=False
            ),
            zaxis=dict(
                title='',
                showgrid=False,
                zeroline=False,
                showticklabels=False,
                backgroundcolor='white',  # Remove grey plane
                showbackground=False
            ),
            bgcolor='white'  # Set overall scene background to white
        ),
        title=f"",
        margin=dict(l=0, r=0, b=0, t=0)
    )


    # Create the figure
    fig = go.Figure(data=[scatter], layout=layout)

    # Camera view for the specified hemisphere
    if hemisphere == 'right':
        fig.update_layout(
            scene_camera=dict(
                eye=dict(x=1.5, y=0, z=0),  # Right view (positive X-axis), dezoom by increasing x
                up=dict(x=0, y=0, z=1),
            )
        )
    elif hemisphere == 'left':
        fig.update_layout(
            scene_camera=dict(
                eye=dict(x=-1.5, y=0, z=0),  # Left view (negative X-axis), dezoom by increasing x
                up=dict(x=0, y=0, z=1),
            )
        )

    # Save the image of the plot
    fig.write_image(output_path)

# Load the vertices and sampled labels for the right hemisphere
right_obj_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_right_surface_smoothed.obj"
right_vertices = load_obj_vertices(right_obj_path)
right_labels = np.loadtxt("/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/right_surface_labels.txt", dtype=int)

# Load the vertices and sampled labels for the left hemisphere
left_obj_path = "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/wm_left_surface_smoothed.obj"
left_vertices = load_obj_vertices(left_obj_path)
left_labels = np.loadtxt("/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/left_surface_labels.txt", dtype=int)

# Create and save 3D plots for the right hemisphere
create_3d_plot(right_vertices, right_labels, 'right', "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/Plotly_right_surface_plot.png")

# Create and save 3D plots for the left hemisphere
create_3d_plot(left_vertices, left_labels, 'left', "/mnt/filerc/LREN/PI_DRAGANSKI/abussy/Neuromorphometrics/Plotly_left_surface_plot.png")
```
